# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

# Read in CSV File

```python
df = pd.read_csv(r"C:\Users\15053\Desktop\Data\company_data.csv")
```

# View Dataset

```python
df.head(20)
```

# Clean Up Market Cap and Revenue Column 

```python
df['Market Cap'] = df['Market Cap'].str.replace('B', '')
df['Revenue'] = df['Revenue'].str.replace('B', '')
df['Market Cap'] = df['Market Cap'].str.replace('M', '')
df['Revenue'] = df['Revenue'].str.replace('M', '')
df['Market Cap'] = df['Market Cap'].str.replace('-', '0')
df['Revenue'] = df['Revenue'].str.replace('-', '0')
df['Market Cap'] = df['Market Cap'].str.replace(',', '')
df['Revenue'] = df['Revenue'].str.replace(',', '')
```

# Rename Market Cap and Revenue Column

```python
df.rename(columns = {'Market Cap': 'market_cap'}, inplace=True)
df.rename(columns = {'Revenue': 'revenue'}, inplace=True)
df.rename(columns = {'Company Name': 'company'}, inplace=True)
df
```

# Switch Market Cap and Revenue Column Data Type from String to Float

```python
df.market_cap = df.market_cap.astype(float)
df.revenue = df.revenue.astype(float)
```

# Filter DF to only Top 10 Companies by Market Cap

```python
df.drop(df[df.market_cap < 626].index, inplace=True)
df
```

# Rename Companies in DF to Stock Ticker Name

```python
df['company'].replace('Microsoft Corporation', 'MSFT', inplace=True)
df['company'].replace('Apple Inc', 'AAPL', inplace=True)
df['company'].replace('NVIDIA Corporation', 'NVDA', inplace=True)
df['company'].replace('Alphabet Inc.', 'GOOG', inplace=True)
df['company'].replace('Amazon.com, Inc.', 'AMZN', inplace=True)
df['company'].replace('Meta Platforms, Inc.', 'META', inplace=True)
df['company'].replace('Berkshire Hathaway Inc.', 'BRK.B', inplace=True)
df['company'].replace('Eli Lilly and Company', 'LLY', inplace=True)
df['company'].replace('Broadcom Inc.', 'AVGO', inplace=True)
df['company'].replace('Taiwan Semiconductor Manufacturing Company Limited', 'TSM', inplace=True)
df
```

# Export Dataset to CSV

```python
df.to_csv('mktg_cap.csv')
```
